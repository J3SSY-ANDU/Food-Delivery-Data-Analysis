```python
weather_df = pd.read_csv('../data/test/jena_climate_2009_2016.csv')
weather_df
```

```python
from copy import deepcopy

weather_df = weather_df[::6*24][:365] # Take one sample every 6 hours for one year
weather_df.reset_index(drop=True, inplace=True)

weather_df_full = deepcopy(weather_df)

weather_df = weather_df[['Date Time', 'p (mbar)', 'T (degC)']]

weather_df
```

```python
weather_df['Date Time'] 
```

```python
weather_df = deepcopy(weather_df)
weather_df['Date Time'] = pd.to_datetime(weather_df['Date Time'], format='%d.%m.%Y %H:%M:%S')

weather_df['Date Time']
```

```python
weather_df.head()
```

```python
sns.lineplot(x='Date Time', y='T (degC)', data=weather_df)
```

```python
g = sns.lineplot(x='Date Time', y='T (degC)', data=weather_df)
g.figure.autofmt_xdate() # Rotate x-axis labels for better readability
```

```python
weather_df_shuffled = weather_df.sample(frac=1) # Shuffle the DataFrame
weather_df_shuffled.reset_index(drop=True, inplace=True) # Reset index after shuffling

weather_df_shuffled
```

```python
g = sns.lineplot(x='Date Time', y='T (degC)', data=weather_df_shuffled) # Plot shuffled data
g.figure.autofmt_xdate()
```

```python
weather_df['above_average_pressure'] = (weather_df['p (mbar)'] > weather_df['p (mbar)'].mean()).astype(int)
weather_df
```

```python
g = sns.lineplot(x='Date Time', y='T (degC)', markers=True, style='above_average_pressure', hue='above_average_pressure', data=weather_df) # Plot with hue based on pressure
g.figure.autofmt_xdate()
```

```python
weather_df['above_average_pressure'].value_counts() # Count of above and below average pressure
```

```python
g = sns.relplot(x='Date Time', y='T (degC)', markers=True, style='above_average_pressure', hue='above_average_pressure', kind='line', col='above_average_pressure', data=weather_df) # Plot with hue based on pressure
g.figure.autofmt_xdate()
```

```python
titanic = sns.load_dataset('titanic') # Load the Titanic dataset
titanic
```

```python
sns.histplot(data=titanic, x='age', bins=30, kde=True, stat='percent', hue='sex') # Histogram with KDE for age distribution
# kde=True adds a kernel density estimate line to the histogram
```

```python
titanic['sex'].value_counts() # Male vs Female count in Titanic dataset
```

```python
sns.barplot(x='sex', y='age', data=titanic) # Bar plot by default shows the average of the data for each category
```

```python
sns.barplot(x='sex', y='age', estimator=sum, data=titanic) # Bar plot with sum as the estimator
```

```python
sns.barplot(y='survived', x='sex', data=titanic) # Bar plot showing survival rate by sex
```

```python
sns.catplot(kind='bar', x='sex', y='survived', hue='class', data=titanic) # Catplot for bar plot of survival rate
```

```python
sns.catplot(kind='bar', x='sex', y='survived', hue='class', col='class', data=titanic) # Catplot with columns for each class
```

```python
import numpy as np
g = sns.histplot(data=titanic, x='class', hue='survived') # Stacked histogram for class and survival
g.set_yticks(np.arange(0, 381, 10))
g.figure.set_size_inches(8, 10)  # width=8, height=6 (in inches)
g.set(title='Survival by Class on Titanic')
```

```python
town_counts = titanic['embark_town'].value_counts()
print(town_counts) # Count of passengers by embarkation town
g = sns.catplot(kind='bar', x='embark_town', y='survived', hue='class', estimator=sum, data=titanic) # Bar plot for embarkation town and survival
g.set(title='Survival by Embarkation Town on Titanic')
```

```python
g = sns.histplot(x='embark_town', hue='class', data=titanic, stat='percent', multiple='stack') # Stacked histogram for embarkation town and class
g.set_yticks(np.arange(0,76,5))
g.figure.set_size_inches(8, 10)  # width=8, height=6 (in inches)
g.set(title='Passenger Distribution by Embarkation Town and Class on Titanic')
```

```python
g = sns.boxplot(x='class', y='fare', hue='sex', data=titanic)
g.set(title='Fare Distribution by Class on Titanic')
g.figure.set_size_inches(6, 12)  # width=8, height=6 (in inches)
g.set_yticks(np.arange(0, 551, 25))
titanic['fare'].describe()
```

```python
outliers_over_500 = titanic[(titanic['fare'] > 500)]['survived'].value_counts()
print(outliers_over_500)  # Display outliers in fare greater than 500

outliers_under_5 = titanic[(titanic['fare'] < 5)]['survived'].value_counts()
print(outliers_under_5)  # Display outliers in fare less than 5
```
