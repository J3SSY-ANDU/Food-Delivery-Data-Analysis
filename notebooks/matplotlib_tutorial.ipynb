### Load necessary libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
x, y = [0,1,2,3,4], [0,2,4,6,8]
# Resize graph (dpi specifies pixels per inch. When saving should use 300 dpi if possible)
plt.figure(figsize=(5, 3), dpi=150) # figsize in inches (width, height), dpi is the resolution in dots per inch

# Line 1

# Keyword argument notation
# plt.plot(x, y, label='2x', color='red', linewidth=2, marker='.', linestyle='--', markersize=10, markeredgecolor='blue') # plot a basic graph

# Shorthand notation
# fmt = '[color][marker][line]'
plt.plot(x, y, 'r.--', label='2x') # check matplotlib plot documentation for more options

# Line 2

# select interval we want to plot points at
x2 = np.arange(0, 4.5, 0.5) # create an array of numbers from 0 to 4.5 with a step of 0.5

# Plot part of the graph as line
plt.plot(x2[:6], x2[:6]**2, 'g', label='x^2') # plot another graph

# Plot part of the graph as dots
plt.plot(x2[5:], x2[5:]**2, 'g--') # plot continuation of the previous graph

# Add a title (specify font parameters with fontdict)
plt.title('My first graph', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20}) # add a title

# X and Y axis labels
plt.xlabel('X-axis') # add x-axis label
plt.ylabel('Y-axis') # add y-axis label

# X and Y tickmarks (scale of your graph)
plt.xticks([0,1,2,3,4]) # change x-axis ticks
plt.yticks([0,2,4,6,8,10]) # change y-axis ticks

# Add a legend
plt.legend() # add a legend

# Save figure (large number of dpi will make the file size larger, but the quality will be better)
plt.savefig('my_graph.png', dpi=300) # save the graph as a PNG file with 300 dpi

# Show plot
plt.show() # we can show the graph if we want to get rid of the default text
```

```python
# we can also use pandas to plot because it has a built-in plot function that uses matplotlib under the hood
df = pd.DataFrame({'x': x, 'y': y})
df.plot(x='x', y='y') # plot a basic graph
plt.show() # we can show the graph if we want to get rid of the default text
```

### Bar Chart

```python
labels = ['A','B','C']
values = [1,4,2]

bars = plt.bar(labels, values) # create a bar graph

# bars[0].set_hatch('/') # set the first bar to have a hatch pattern
# bars[1].set_hatch('o') # set the second bar to have a hatch pattern
# bars[2].set_hatch('*') # set the third bar to have a hatch pattern

patterns = ['/', 'o', '*'] # create a list of hatch patterns
for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.figure(figsize=(5, 3), dpi=150) # resize the graph

plt.show() # we can show the graph if we want to get rid of the default text
```

```python
gas = pd.read_csv('../data/test/gas_prices.csv') # read the CSV file

plt.figure(figsize=(5,3), dpi=150) # resize the graph

plt.title('Gas Prices Over Time (in USD)', fontdict={'fontweight': 'bold', 'fontsize': 14}) # add a title

# plt.plot(gas.Year, gas.USA, 'b.-', label='USA') # plot the USA gas prices
# plt.plot(gas.Year, gas.Canada, 'r.-', label='Canada') # plot the Canada gas prices
# plt.plot(gas.Year, gas['South Korea'], 'g.-', label='South Korea') # plot the South Korea gas prices
# plt.plot(gas.Year, gas.Australia, 'y.-', label='Australia') # plot the Australia gas prices

# Plot all countries in one line
plt.plot(gas.Year, gas[['USA', 'Canada', 'South Korea', 'Australia']], marker='.', markersize=10)

# Plot all countries in the dataset
for country in gas.columns[1:]:
    plt.plot(gas.Year, gas[country], marker='.', label=country)

plt.xticks(gas.Year[::3].tolist() + [2011]) # change x-axis ticks to show every 3rd year

plt.xlabel('Year') # add x-axis label
plt.ylabel('Gas Price (USD)') # add y-axis label

# bbox_to_anchor and loc are parameters for positioning the legend in matplotlib.
# 'loc' specifies the anchor point of the legend box (e.g., 'upper left', 'lower right', etc.).
# 'bbox_to_anchor' specifies the (x, y) coordinates to place the anchor point, relative to the axes.
# For example, bbox_to_anchor=(1.05, 1) with loc='upper left' places the legend just outside the top right of the plot.
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left') # place legend outside the plot
plt.tight_layout() # adjust layout to make room for the legend

# I had to change the title size and add bbox_to_anchor, loc and tight_layout to make the legend fit in the graph so it could be seen in the save file
plt.savefig('gas_prices.png', dpi=300) # save the graph as a PNG file with 300 dpi

plt.show() # show the graph
```

### Load fifa data

```python
fifa = pd.read_csv('../data/test/fifa_data.csv') # read the CSV file

fifa.head()
```

### Histograms

```python
bins = [40, 50, 60, 70, 80, 90, 100] # create bins for the histogram

plt.hist(fifa.Overall, bins=bins, color='#abcdef') # create a histogram of the Overall column

plt.xticks(bins) # change x-axis ticks to show the bins

plt.ylabel('Number of Players') # add y-axis label
plt.xlabel('Skill Level') # add x-axis label

plt.title('Distribution of Player Skills in FIFA 2018') # add a title

plt.show() # show the graph
```

### Pie Charts

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].shape[0] # filter the dataframe for left-footed players
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].shape[0] # filter the dataframe for right-footed players

plt.pie([left, right], labels=['Left', 'Right'], colors=['#abcdef', '#123456'], autopct='%.2f%%')

plt.title('Foot Preference of FIFA Players') # add a title

plt.savefig('fifa_foot_preference.png', dpi=300) # save the graph as a PNG file with 300 dpi

plt.show() # show the graph
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight] # convert the Weight column to integers

light = fifa.loc[fifa.Weight < 125].shape[0] # filter the dataframe for players under 125 lbs
medium_light = fifa.loc[fifa.Weight.between(125, 150)].shape[0] # filter the dataframe for players between 125 and 150 lbs
medium = fifa.loc[fifa.Weight.between(150, 175)].shape[0] # filter the dataframe for players between 150 and 175 lbs
medium_heavy = fifa.loc[fifa.Weight.between(175, 200)].shape[0] # filter the dataframe for players between 175 and 200 lbs
heavy = fifa.loc[fifa.Weight > 200].shape[0] # filter the dataframe for players over 200 lbs

weights = [light, medium_light, medium, medium_heavy, heavy] # create a list of weights
labels = ['< 125 lbs', '125-150 lbs', '150-175 lbs', '175-200 lbs', '> 200 lbs'] # create a list of labels
colors = ['#abcdef', '#123456', '#654321', '#fedcba', '#987654'] # create a list of colors

explode = (0.4, 0.1, 0.1, 0.1, 0.4)
plt.pie(weights, labels=labels, colors=colors, autopct='%.2f%%', pctdistance=0.8, explode=explode) # create a pie chart of the weights
plt.title('Weight Distribution of FIFA Players') # add a title

plt.savefig('fifa_weight_distribution.png', dpi=300) # save the graph as a PNG file with 300 dpi

plt.show() # show the graph
```
